```python
def fun4(a):
    m=1
    for i in range(1,a+1):
        x=lambda n: 1 if n<=1 else n*x(n-1)
        m*=x(i)
    return m
```

```python
fun4(6)
```

```python
def fun6(a,b,c):
    if c=='+':
        return a+b
    if c=='-':
        return a-b
    if c=='*':
        return a*b
    if c=='/':
        return a/b
```

```python
fun6(5,2,'*')
```

```python
def fun7(a):
    c=''
    for i in range(1,a+1):
        if i>1:
            c+=','
        if i%2!=0:
            c+='"I CAN"'
        else:
            c+='"I WILL"'
    return c
```

```python
print(fun7(7))
```

```python
def fun8(a):
    s=set(a)
    p=0
    for i in s:
        p+=a.count(i)//2
    return p
```

```python
fun8([1,2,1,2,1,3,2])
```

```python
def fun9(a,b):
    c=set(a+b)
    d=max(c)-min(c)
    if d+1==len(c):
        return True
    return False
```

```python
fun9([7,4,5,1],[2,3,6])
```

```python
fun9([1,4,6,5],[2,7,8,9])
```
